```python
# Create a 4x3 matrix of random integers
matrixB = np.random.randint(0, 10, size=(4, 3))  # Generates random integers between 0 and 99 (inclusive)

print(matrixB)
print(matrixB.shape)
```

```python
print(matrixB)
col_indices=np.array([0,1,2])
fil_indices=np.array([1,2])

for fil,col in zip(fil_indices,col_indices):
  # print(fil," - ",col)
  print(matrixB[col,fil])
print(col_indices)
print(fil_indices)
```

```python
mask=matrixB>7
print(mask)
```

```python
print(matrixB[mask])
```

```python
print(matrixB[matrixB>7])
```

```python
matrixB[matrixB % 2 == 0] += 100

matrixB
```

```python
filter=(matrixB>3) & (matrixB<9)
matrixB[filter]+=100
print(matrixB)
```
